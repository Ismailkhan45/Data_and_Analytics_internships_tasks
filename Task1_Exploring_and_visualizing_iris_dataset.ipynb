```python
plt.figure(figsize=(14,7))
plt.subplot(2,2,1)
sns.scatterplot(data = df, x='sepal_length', y='sepal_width', hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.subplot(2,2,2)
sns.scatterplot(data = df, x="petal_length", y="petal_width", hue="species")
plt.title("Petal_Length vs Petal_Width")
plt.tight_layout()
plt.show()
```

Scatter Plot Observation
In these Scatter plot we can say the Relationsip between Sepal_Lenght vs Sepal_Width is Moderate relationship. This is describing that sepal measurements alone may not clearly distinguish flower categories.

In the 2nd chart Petal_Length vs Petal_Width demonstrates a strong and positive relationship where as Petal_Length increase Petal_Width also increase. Distinct clustering among species is visible suggesting that petal measurements are more effective for differentiating iris flower species

```python
plt.figure(figsize=(13,5))
sns.scatterplot(data = df, x='sepal_length', y='petal_length' , hue='species')
plt.title("Speal_Lenght vs Petal_Lenght Obesrvation ")
plt.show()
```

```python
plt.figure(figsize=(13,7))   # Set Size of Picture 
plt.subplot(2,2,1)
sns.boxplot(df['sepal_length'])
plt.title("Sepal Length")

plt.subplot(2,2,2)
sns.boxplot(df['sepal_width'])
plt.title("Sepal Width")

plt.subplot(2,2,3)
sns.boxplot(df["petal_length"])
plt.title("Petal Length")

plt.subplot(2,2,4)
sns.boxplot(df["petal_width"])
plt.title("Petal Width")

plt.tight_layout()
plt.show()
```

Box Plot Observation
BoxPlot Analysis Indicate there are few outlier in the Dataset in Seapl_Width Column. These outlier are very small number of so we don't need to remove then our dataset is clean and useable for Analysis

```python
plt.figure(figsize=(13,7))
plt.subplot(2,2,1)
sns.histplot(df["sepal_length"], kde=True)
plt.title("Distribution of Sepal Length")

plt.subplot(2,2,2)
sns.histplot(df["sepal_width"], kde=True)
plt.title("Distribution of Sepal Width")

plt.subplot(2,2,3)
sns.histplot(df["petal_length"], kde=True)
plt.title("Distribution of Petal Lenth ")

plt.subplot(2,2,4)
sns.histplot(df["petal_width"], kde=True)
plt.title("Distribution of Petal Width")

plt.tight_layout()
plt.show()
```
